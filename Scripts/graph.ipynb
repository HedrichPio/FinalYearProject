```python
import os
import numpy as np
from datetime import datetime
from matplotlib import pyplot as plt, dates as mdates
#assign directory

no_clouds_xyz_dir = 'Documents/UNI/PlanetData_FYP/OutputFilesYala/TempFiles'

 
list_of_xyz_files = os.listdir(no_clouds_xyz_dir)

#open first file
xyz_file = open(no_clouds_xyz_dir+"/"+list_of_xyz_files[0], "r")

#read all lines and put to a list
xyz_line_list = xyz_file.readlines()


for i in range(len(xyz_line_list)):

#array of ndvi values for each day
    ndvi_values_for_each_day = []


    for j in range(len(list_of_xyz_files)):
        
    #open xyz files of ndvi and clouds
        no_cloud_ndvi_xyz_file = open(no_clouds_xyz_dir+"/"+list_of_xyz_files[j], "r")
        
    #add all xyz lines as list
        no_clouds_ndvi_line_list = no_cloud_ndvi_xyz_file.readlines()

        ndvi_line = no_clouds_ndvi_line_list[i].split(" ")

        floatnum = float(ndvi_line[2])

        roundnum = round(floatnum,2)

        ndvi_values_for_each_day.append(roundnum)

  

    dates = ["04/04/2022","23/04/2022", "29/04/2022", "30/05/2022","27/06/2022","02/07/2022","08/07/2022","19/07/2022","21/07/2022","28/07/2022","07/08/2022","15/08/2022","24/09/2022"]
    
    x = [datetime.strptime(d, "%d/%m/%Y").date() for d in dates]

    y = ndvi_values_for_each_day

    ax = plt.gca()

    ax.xaxis.set_major_formatter(mdates.DateFormatter("%d-%m-%Y"))

    
    plt.rcParams["figure.figsize"] = (15,15)

    # naming the x axis
    plt.xlabel('days')

    # naming the y axis
    plt.ylabel('ndvi')
      
    # title of graph
    plt.title('NDVI in pixel no. '+ str(i))

    # plot
    ax.plot(x, y) 
    
    # show the plot
    plt.show()

```
